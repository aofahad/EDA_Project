```python
#Import the libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read in the Desktop(cars.csv) csv called cars_price.
cars_prices = pd.read_csv('cars.csv')
cars_prices
```

```python
#The numbers of rows & columns in the Dataset.
cars_prices.shape
```

```python
#This method to know 5 of rows and columns by default.
cars_prices.head()
```

```python
#Print a concise summary of data.
cars_prices.info()
```

```python
#checking the null values.
cars_prices.isna().sum()
```

```python
#Drob null values in rows.
cars_prices.dropna(inplace = True)
cars_prices.isna().sum()
```

```python
#Here we replace all out outliers in columns. Hint: df['column name'].unique() >> to check outliers.
cars_prices['Year'] = cars_prices['Year'].replace(4,2010)
cars_prices['Engine HP'] = cars_prices['Engine HP'].replace([1001,4,750,731,720,707],320)
cars_prices['Engine Cylinders'] = cars_prices['Engine Cylinders'].replace([0,10,12,16],8)
cars_prices['Number of Doors'] = cars_prices['Number of Doors'].replace(3,4)
cars_prices['Popularity'] = cars_prices['Popularity'].replace([2,21,26,86,67,61],100)
cars_prices['MSRP'] = cars_prices['MSRP'].replace(4,50000)
```

```python
#This mehod to delete unnecessary some columns as you see below.
cars_prices.drop(['highway MPG','city mpg','Engine Fuel Type','Driven_Wheels','Vehicle Size'], axis = 1, inplace = True)
```

```python
#To minimeis the zeroes after the comma .0000 ....
pd.set_option('display.float_format', lambda x:'%.2f'% x)
```

```python
#Used to view some basic statistical details like percentile, mean, std.
cars_prices.describe()
```

```python
#We write this code to see which type of car is more popular.
plt.figure(figsize=[14,8])
sns.histplot(data=cars_prices, x="Make", kde=True)
plt.xticks(fontsize = 13 , rotation=90)
plt.yticks(fontsize = 13)
plt.xlabel('NAME OF CARS')
plt.ylabel('COUNT')
plt.title('The Popular Car',fontsize=20);
plt.savefig('The Popular.png')
```

```python
#This code to show sales of car style.
plt.figure(figsize=[14,8])
sns.set_theme(style="darkgrid")
sns.countplot(x="Vehicle Style", data=cars_prices)
plt.xticks(fontsize = 13, rotation = 90)
plt.yticks(fontsize = 13)
plt.xlabel('VEHICLE STYLE')
plt.ylabel('COUNT')
plt.title('The Sales Of Car Style',fontsize=20)
plt.grid();
plt.savefig('The Sales Of Car Style.png')
```

```python
#This code to see cars that has the 300 of engine horsepower.
f = cars_prices[cars_prices['Engine HP'] == 300]
plt.figure(figsize=[14,8])
sns.barplot(f['Make'],cars_prices['Engine HP'], palette="Purples_d")
plt.xticks(fontsize = 13, rotation = 45)
plt.yticks(fontsize = 13)
plt.xlabel('CARS')
plt.ylabel('ENGINE HP')
plt.title('Cars That has 300 Engine HP',fontsize=20)
plt.grid();
plt.savefig('Cars That has 300 Engine HP.png')
```

```python
#This final code to show how many engine cylinders of each car we have.
z=cars_prices.groupby('Make')[['Engine Cylinders']].max().sort_values(by=['Engine Cylinders'], ascending=False)
plt.figure(figsize=[14,8])
plt.bar(z.index,z['Engine Cylinders'] , color='r')
plt.xticks(fontsize = 13 , rotation = 90)
plt.yticks(fontsize = 13)
plt.xlabel('NAME OF CARS')
plt.ylabel('NUMBER OF ENGINE CYLINDERS')
plt.title('Engine Cylinders Of The Cars', fontsize=20);
plt.savefig('Engine Cylinders Of The Cars.png')
```
